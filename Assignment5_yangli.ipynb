```python
#import the necessary libraries
import pandas as pd
import requests
import matplotlib.pyplot as plt
import numpy as np
```

```python
#get data from the csv
smoking_df = pd.read_csv('/Users/liyang/Desktop/Collectingdata_resit/smoking.csv', delimiter=",", encoding='utf-8')
```

```python
#drop of the rows that we do not need
smoking_df = smoking_df.drop(smoking_df[smoking_df['smoke'] == 'No'].index) 
```

```python
#check data
smoking_df.sample(20)
#data is ready for the next step
```

```python
#dividing artists by marital_status
Divorced_filter = smoking_df['marital_status'] == 'Divorced'
Divorced_smoking_df = smoking_df[Divorced_filter]
Married_filter = smoking_df['marital_status'] == 'Married'
Married_smoking_df = smoking_df[Married_filter]
Single_filter = smoking_df['marital_status'] == 'Single'
Single_smoking_df = smoking_df[Single_filter]
Seperated_filter = smoking_df['marital_status'] == 'Seperated'
Seperated_smoking_df = smoking_df[Seperated_filter]
Widowed_filter = smoking_df['marital_status'] == 'Widowed'
Widowed_smoking_df = smoking_df[Widowed_filter]
```

```python
#saving plot
plot_x = ('Divorced','Married','Single','Seperated','Widowed')
plot_y = [sum(Divorced_filter),sum(Married_filter),sum(Single_filter),sum(Seperated_filter),sum(Widowed_filter)]
plt.bar(plot_x,plot_y)
plt.title('Comparison of smokers marital_status in the UK')
plt.show
```

Conclusion and Reflection

It is clear from the results that there are still some differences in the choice of whether to smoke or not among people with different marital status. Married and single people are the majority of smokers in the UK, with the largest number of single smokers, and the number of divorced and widowed smokers is significantly less than that of married and single people. One of the things that surprised me was that the number of seperated smokers was actually 0.

This may have something to do with the bias in the sample collected. In the future, we can expand the sample size. Also, collect samples of smokers from more countries. In this way, the bias is reduced.
