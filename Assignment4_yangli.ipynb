```python
#import the necessary libraries
import pandas as pd
import requests
import matplotlib.pyplot as plt
import numpy as np
```

```python
#get data from the csv
Artists_df = pd.read_csv('/Users/liyang/Desktop/Collectingdata_resit/artist_data.csv', delimiter=",", encoding='utf-8')
```

```python
#drop of the rows that we do not need
Artists_df = Artists_df.drop(Artists_df[Artists_df['yearOfBirth'] > 1900].index) 
Artists_df = Artists_df.drop(Artists_df[Artists_df['yearOfBirth'] < 1800].index) 
```

```python
#check data
Artists_df.sample(20)
#data is ready for the next step
```

```python
#dividing artists by gender
women_filter = Artists_df['gender'] == 'Female'
women_Artists_df = Artists_df[women_filter]
men_filter = Artists_df['gender'] == 'Male'
men_Artists_df = Artists_df[men_filter]
```

```python
#saving plot
plot_x = ('women','men')
plot_y = [sum(women_filter),sum(men_filter)]
plt.bar(plot_x,plot_y)
plt.title('Comparison of the number of male and female artists in the 19th century')
plt.show
```

Conclusion and Reflection

After filtering and sifting through the dataset, we can find that: in this dataset, there is an imbalance in the gender ratio of 19th century artists, with the number of male artists being much larger than that of female artists. This may be related to the prejudice against women in the art world (Anna Brady). In the last 10 years, only six women's art has had annual auction sales that could be ranked in the top 100 in terms of sales. However, according to the 2020 annual reports of Christie's, Sotheby's and Phillips, there is indeed an increase in demand for works by female artists in the market today. Major auction houses are also now publicly stating that they are reassessing and revaluing women artists, which has the potential to ultimately change the way artists are differentiated by gender and bring profound changes to the contemporary art market.

However, this dataset does not collect the artists' nationalities, only their places of birth and death. This may not allow for further study of the artists' nationality backgrounds and whether this also creates a gender imbalance. Therefore, in the future, we can collect the nationality of artists and use it to study whether there are also large differences between male and female artists in different regions.
